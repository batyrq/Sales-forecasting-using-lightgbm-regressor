```python
posoil = {}
for i in range(len(oil)):
    posoil[oil.loc[i,'date']] = oil.loc[i,'dcoilwtico']
```

```python
oil
```

```python
train
```

```python
for i in range(len(train)):
    if i % 100000 == 0:
        print(i)
    if train.loc[i,'date'] in posoil:
        train.loc[i,'dcoilwtico'] = posoil[train.loc[i,'date']]
    else:
        train.loc[i,'dcoilwtico'] = np.mean(oil['dcoilwtico'])
train
```

```python
for i in range(len(test)):
    if i % 100000 == 0:
        print(i)
    if test.loc[i,'date'] in posoil:
        test.loc[i,'dcoilwtico'] = posoil[test.loc[i,'date']]
    else:
        test.loc[i,'dcoilwtico'] = np.mean(oil['dcoilwtico'])
test
```

```python
import lightgbm as lgb

model = lgb.LGBMRegressor(
    n_estimators=1000,
    learning_rate=0.05,
    num_leaves=64,
    random_state=42
)
y_train = train['sales']
X_train = train.drop(['sales','date'],axis=1)

X_train
```

```python
model.fit(X_train,y_train)
```

```python
sample_submission = pd.read_csv('/kaggle/input/store-sales-time-series-forecasting/sample_submission.csv')
sample_submission
```

```python
test = test.drop('date',axis = 1)
sample_submission['sales'] = model.predict(test)
for i in range(len(sample_submission)):
    sample_submission.loc[i,'sales'] = max(sample_submission.loc[i,'sales'],0)
```

```python
sample_submission.to_csv('submission.csv',index = False)
```
